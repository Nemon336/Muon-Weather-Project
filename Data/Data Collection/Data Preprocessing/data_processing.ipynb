```python
import pandas as pd
import re
import numbers
import numpy as np

# Load the CSV file
df = pd.read_csv('weather_data.csv', encoding='utf-8')  # Adjust the file name and path as needed
# print(df.head())

unique_elements = df['weather'].nunique()
print(unique_elements)

# display all unique elements
# print(df['weather'].unique())

# Function to map Chinese days to numbers
def map_day_to_num(day):
    mapping = {'星期一': 1, '星期二': 2, '星期三': 3, '星期四': 4, '星期五': 5, '星期六': 6, '星期日': 7}
    return mapping.get(day, None)  # Returns None if the day is not found

# Function to remove non-numeric characters (e.g., %) and convert to float
def clean_numeric_field(value):
    if isinstance(value, numbers.Number):
        return value
    numeric_part = re.sub(r'[^0-9.]+', '', value)
    return float(numeric_part) if numeric_part else None

# Function to map weather conditions to numerical/categorical values
def map_weather_to_num(weather):
    weather_mapping = {'晴': 1, '阴': 2, '多云': 3, '小雨': 4, '霾': 5, '雾': 6, '中雪': 7, '雪': 8, '小雪': 9} 
    return weather_mapping.get(weather, None)  # Returns None if the weather condition is not found

# Applying the functions to the DataFrame
df['week'] = df['week'].apply(map_day_to_num)
df['humidity'] = df['humidity'].apply(clean_numeric_field)
df['pressure'] = df['pressure'].apply(clean_numeric_field)
df['weather'] = df['weather'].apply(map_weather_to_num)
df['temperature'] = df['temperature'].apply(clean_numeric_field)
df['city'] = df['city'].replace({np.nan: '海淀'}) # Replace NaN with '海淀'

# Display the first few rows of the DataFrame to check the changes
df.head()

# Save the cleaned data to a new CSV file
df.to_csv('cleaned_weather_data.csv', index=False, encoding='utf-8')  # Adjust the file name and path as needed
```
